```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import BernoulliNB
```

```python
#import dataset

data = pd.read_csv('/content/Youtube01-Psy.csv')
data.head()
```

```python
#splitting
data = data[['CONTENT','CLASS']]
data.head()
```

```python
#map columns
# 0- not spam
# 1- spam

data['CLASS'] = data['CLASS'].map({0:"Not Spam", 1:"Spam"})
data.head()
```

```python
data[0]
```

```python
x = data["CONTENT"].to_numpy()
y = data["CLASS"].to_numpy()

cv = CountVectorizer()
x = cv.fit_transform(x)
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.2, random_state = 42 )
```

```python
model = BernoulliNB()
model.fit(xtrain, ytrain)

print(model.score(xtest, ytest))
```

```python
#validate the model

#sample= "Hello Everyone"
#sample = "i turned it on mute as soon is i came on i just wanted to check the  views...ï»¿"
sample = "Subscribe to my channel ï»¿"

data = cv.transform([sample]).toarray()

print(model.predict(data))
```
